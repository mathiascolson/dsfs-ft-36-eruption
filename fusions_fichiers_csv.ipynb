```python
import os
import pandas as pd
import glob

# 1. DÉFINIR LE RÉPERTOIRE
# Si le script est dans le dossier 'PROJET_COLLECTIFS' ou si vous l'exécutez
# depuis ce dossier, utilisez '.' (point).
chemin_du_repertoire = '.' 

# Si le répertoire est ailleurs, modifiez comme suit :
# chemin_du_repertoire = 'votre/chemin/vers/PROJET_COLLECTIFS'

# 2. TROUVER TOUS LES FICHIERS CSV
# Ceci va trouver les 3 fichiers que vous avez montrés.
motif_fichiers = os.path.join(chemin_du_repertoire, "*.csv")
tous_les_fichiers = glob.glob(motif_fichiers)

# 3. CONCATÉNATION DES DONNÉES
liste_dataframes = []

print(f"Début de la lecture des {len(tous_les_fichiers)} fichiers CSV...")

for nom_fichier in tous_les_fichiers:
    # Lecture du fichier et ajout à la liste
    df = pd.read_csv(nom_fichier)
    liste_dataframes.append(df)
    print(f"-> Fichier lu : {os.path.basename(nom_fichier)} (lignes : {len(df)})")

# 4. FUSION ET SAUVEGARDE
if liste_dataframes:
    # Concaténation des DataFrames (empilement vertical)
    dataframe_final = pd.concat(liste_dataframes, ignore_index=True)
    
    # Nom du fichier de sortie
    nom_fichier_sortie = 'DONNEES_FUSIONNEES_COMPLETES.csv'
    
    # Écriture du fichier final, sans l'index Pandas
    dataframe_final.to_csv(nom_fichier_sortie, index=False, encoding='utf-8')
    
    print("---")
    print(f"✅ Opération réussie !")
    print(f"Taille totale du fichier fusionné : {len(dataframe_final)} lignes.")
    print(f"Le résultat est enregistré sous : **{os.path.abspath(nom_fichier_sortie)}**")

else:
    print("\n⚠️ Aucun fichier CSV trouvé correspondant au motif.")
```
